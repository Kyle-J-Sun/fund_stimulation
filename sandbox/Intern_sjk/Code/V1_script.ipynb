```python
def get_netasset(account_name,net_value,cash_rela,df_hs):
    '''

    INPUT：
    account_name(str) 账户代码，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    
    OUTPUT：
    tmp(df) 调整后净值

    '''
    tmp = net_value[net_value['vc_fund_code']==account_name]
    tmp.reset_index(drop=True,inplace=True)
    tmp.set_index('d_date',inplace=True)
    tmp = tmp.sort_index()
    tmp = tmp.reindex(df_hs.index)
    cash_tmp = cash_rela[cash_rela['vc_gp3_account']==account_name].copy()
    tmp = tmp.merge(cash_tmp,left_index=True,right_on='trade_dt',how='outer')
    tmp.drop('vc_gp3_account',axis=1,inplace=True)
    
    #cash_rela[cash_rela['vc_gp3_account']==account_name]
    tmp['cashflow'] = tmp['cashflow'].fillna(0)
    tmp['raw_f_index_value'] = tmp['f_index_value']
    tmp.reset_index(drop=True,inplace=True)
    cash_index = list(tmp[tmp['cashflow']>0].index)
    for i in range(len(cash_index)):
        if np.isnan(tmp.loc[cash_index[i],'f_index_value']):
            tmp.loc[cash_index[i],'vc_fund_code'] = tmp.loc[cash_index[i]+1,'vc_fund_code']
            tmp.loc[cash_index[i],'vc_indicator_code'] = tmp.loc[cash_index[i]+1,'vc_indicator_code']
            tmp.loc[cash_index[i],'vc_indicator_name'] = tmp.loc[cash_index[i]+1,'vc_indicator_name']
            tmp.loc[cash_index[i],'f_index_value'] = 0
            tmp.loc[cash_index[i],'raw_f_index_value'] = 0
        #if tmp.loc[cash_index[i]+1,'f_index_value']/tmp.loc[cash_index[i],'cashflow']<0.5:
        #单独修复
        tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value'] = tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value']+tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'cashflow']
        if (tmp.loc[cash_index[i],'f_index_value'])/(tmp.loc[cash_index[i]+2,'f_index_value']+0.0001)<0.9:
            cash_missed = tmp.loc[cash_index[i],'cashflow']
            j =cash_index[i]
            #3-一般不会连续三天未增加
            while (tmp.loc[j,'f_index_value'])/(tmp.loc[j+2,'f_index_value']+0.0001)<0.9:
                #单独修复的账户
                if (tmp.loc[j,'vc_fund_code'] not in ['LH14G2','LH14H2','LH19I2'])|(tmp.loc[j,'trade_dt'] not in [dt.datetime(2020,9,23),dt.datetime(2020,9,24),dt.datetime(2018,10,16),dt.datetime(2019,4,15)]):
                    tmp.loc[j,'f_index_value']=tmp.loc[j,'f_index_value']+cash_missed
                j = j+1   
    tmp = tmp.dropna(axis=0,how='any')
    tmp.reset_index(drop=True,inplace=True)
    #tmp = tmp.set_index('trade_dt')
    tmp = tmp[['trade_dt','vc_fund_code','f_index_value','vc_indicator_code','vc_indicator_name']].copy()
    tmp.columns = ['d_date','vc_fund_code','f_index_value','vc_indicator_code','vc_indicator_name']
    tmp = tmp.sort_values('d_date')
    
    if account_name in ['LH0026','LH19C2','LH19E2','LH19D2']:
        tmp  = tmp[tmp['d_date']>=dt.datetime(2019,5,16)]
    tmp.reset_index(drop=True,inplace=True)
    
    return tmp
```

```python
#净值系列函数
#利用现金流表作为现金流进行净值计算
def get_netvalue(account_name,net_value,cash_rela,df_hs,an_rate):
    '''

    获取单位净值，剔除现金流影响
    
    INPUT：
    account_name(str) 账户代码，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs(df) 沪深300数据
    
    OUTPUT：
    compare(df) 计算净值及沪深300全收益

    '''
    tmp = net_value[net_value['vc_fund_code']==account_name]
    tmp.reset_index(drop=True,inplace=True)
    tmp.set_index('d_date',inplace=True)
    tmp = tmp.sort_index()
    tmp = tmp.reindex(df_hs.index)
    cash_tmp = cash_rela[cash_rela['vc_gp3_account']==account_name].copy()
    tmp = tmp.merge(cash_tmp,left_index=True,right_on='trade_dt',how='outer')
    tmp.drop('vc_gp3_account',axis=1,inplace=True)
    tmp['cashflow'] = tmp['cashflow'].fillna(0)
    tmp['raw_f_index_value'] = tmp['f_index_value']
    tmp.reset_index(drop=True,inplace=True)
    cash_index = list(tmp[tmp['cashflow']>0].index)
    if account_name != 'LHOTRE':
        for i in range(len(cash_index)):
            if np.isnan(tmp.loc[cash_index[i],'f_index_value']):
                tmp.loc[cash_index[i],'vc_fund_code'] = tmp.loc[cash_index[i]+1,'vc_fund_code']
                tmp.loc[cash_index[i],'vc_indicator_code'] = tmp.loc[cash_index[i]+1,'vc_indicator_code']
                tmp.loc[cash_index[i],'vc_indicator_name'] = tmp.loc[cash_index[i]+1,'vc_indicator_name']
                tmp.loc[cash_index[i],'f_index_value'] = 0
                tmp.loc[cash_index[i],'raw_f_index_value'] = 0
                
            #单独修复
            tmp_updt = tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value']
            tmp_updt = tmp_updt + tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'cashflow']
            tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value'] = tmp_updt
            
            if (tmp.loc[cash_index[i],'f_index_value'])/(tmp.loc[cash_index[i]+2,'f_index_value']+0.0001)<0.9:
                cash_missed = tmp.loc[cash_index[i],'cashflow']
                j =cash_index[i]
                #3-一般不会连续三天未增加
                while (tmp.loc[j,'f_index_value'])/(tmp.loc[j+2,'f_index_value']+0.0001)<0.9:
                    #单独修复的账户
                    if (tmp.loc[j,'vc_fund_code'] not in ['LH14G2','LH14H2','LH19I2'])|(tmp.loc[j,'trade_dt'] not in [dt.datetime(2020,9,23),dt.datetime(2020,9,24),dt.datetime(2018,10,16),dt.datetime(2019,4,15)]):
                        tmp.loc[j,'f_index_value']=tmp.loc[j,'f_index_value']+cash_missed
                    j = j+1   
    tmp = tmp.dropna(axis=0,how='any')
    tmp.reset_index(drop=True,inplace=True)
    tmp['mod_return'] = 1
    for i in range(1,len(tmp.index)):
        tmp.loc[tmp.index[i],'mod_return'] = (tmp.loc[tmp.index[i],'f_index_value']- \
                                              tmp.loc[tmp.index[i],'cashflow'])/ \
                                            tmp.loc[tmp.index[i-1],'f_index_value']
    tmp['netvalue'] = np.cumprod(tmp['mod_return'])
    tmp = tmp.set_index('trade_dt')
    compare = pd.concat([df_hs,tmp['netvalue']],axis=1)
    compare = compare.dropna(axis=0,how='any')
    compare['s_dq_close'] = compare['s_dq_close']/compare['s_dq_close'][0]
    compare.columns = ['hs300','hs300_return','netvalue']
    
    if account_name in ['LH0026','LH19C2','LH19E2','LH19D2']:
        compare  = compare.loc[dt.datetime(2019,5,16):]
        compare['hs300'] = compare['hs300']/compare.loc[dt.datetime(2019,5,16),'hs300']
        compare['netvalue'] = compare['netvalue']/compare.loc[dt.datetime(2019,5,16),'netvalue']
    
    time_idx = pd.date_range(compare.index[0],compare.index[-1],freq='D')
    #time_idx = compare.index
    daily_rate = (1+an_rate)**(1/365)
    AbsoGrowth = pd.DataFrame({'AbsoGrowth':[1 for i in range(len(time_idx))]},index=time_idx)
    for i in range(len(AbsoGrowth)-1):
        AbsoGrowth.loc[AbsoGrowth.index[i+1],'AbsoGrowth'] = AbsoGrowth.loc[AbsoGrowth.index[i],'AbsoGrowth']*daily_rate    
    AbsoGrowth = AbsoGrowth.reindex(compare.index)
    compare['absolute'] = AbsoGrowth['AbsoGrowth']

    
    return compare

def plot_net_value(compare,an_rate,category):
    '''
    
    画净值曲线
    
    INPUT：
    compare(df) 计算净值及沪深300全收益，
    an_rate(flaot) 绝对收益率，
    category(str) 账户类型，rela(相对收益),absol(绝对收益)
    
    OUTPUT：
    无返回值

    '''

    fig = plt.figure(figsize=(13,4))
    ax1 = fig.add_subplot(111)
    fig1 = ax1.plot(compare['netvalue'],'b-',label='Net Value')
    fig2 = ax1.plot(compare['hs300'],'r-',label='HS300')
    if category == 'absol':
        ax1.plot(compare['absolute'],'g-',label=str(format((an_rate*100),'.2f'))+'%-Absolute Target')
        ax2 = ax1.twinx()
        #ax2.plot(compare['netvalue']-compare['absolute'],color = "gray",alpha = 0.5,label='Excess Return')
        fig3 = ax2.fill_between(compare.index,compare['netvalue']-compare['absolute'],color = "gray",alpha = 0.5,label='Excess Return')
    else:
        ax2 = ax1.twinx()
        #ax2.plot(compare['netvalue']-compare['hs300'],color = "gray",alpha = 0.5,label='Excess Return')
        fig3 = ax2.fill_between(compare.index,compare['netvalue']-compare['hs300'],color = "gray",alpha = 0.5,label='Excess Return')

    plt.xticks(rotation=45)
    ax1.set_ylabel('Net Value')
    ax2.set_ylabel('Excess Return')
    ax1.set_xlabel('Time')
    plt.title('Net Value')
    fig.legend(loc=0,bbox_to_anchor=(0.2,1),bbox_transform=ax1.transAxes,frameon=0)
    plt.grid()
    return None

def plot_position(compare,an_rate,net_value,cash_rela,df_hs,df_stock,account_name):
    '''
    
    画绝对收益账户净值曲线及仓位
    
    INPUT：
    compare(df) 计算净值及沪深300全收益，
    an_rate(flaot) 绝对收益率，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs(df) 沪深300数据,
    df_stock(df) 持仓数据,
    account_name(str) 账户代码
    
    OUTPUT：
    无返回值

    '''
    plt.style.use('ggplot')
    position_tmp = get_position(account_name,net_value,cash_rela,df_hs,df_stock)
    compare = compare.merge(position_tmp,left_index=True,right_on='d_date',how='left')
    compare = compare.set_index('d_date')
    fig = plt.figure(figsize=(13,4))
    ax1 = fig.add_subplot(111)
    fig1 = ax1.plot(compare['netvalue'],'b-',label='Net Value')
    fig2 = ax1.plot(compare['hs300'],'r-',label='HS300')
    fig3 = ax1.plot(compare['absolute'],'g-',label=str(format((an_rate*100),'.2f'))+'%-Absolute Target')
    ax2 = ax1.twinx()
    #ax2.plot(compare['position'],color = "k",alpha = 0.4,label='Position')
    fig4 = ax2.fill_between(compare.index,compare['position'],color = "k",alpha = 0.4,label='Position')

    plt.xticks(rotation=45)
    ax1.set_ylabel('Net Value')
    ax2.set_ylabel('Position')
    ax1.set_xlabel('Time')
    plt.title('Net Value')
    fig.legend(loc=0,bbox_to_anchor=(0.25,1),bbox_transform=ax1.transAxes,frameon=0)
    plt.grid()
    return None

def get_drawdown(compare,column):
    '''

    INPUT：
    compare(df) 计算净值及沪深300全收益，
    column(str) 计算最大回撤的列
    
    OUTPUT：
    min(dd_tmp_list)(float) 最大回撤
    
    '''
    
    dd_tmp_list = []
    for i in range(len(compare)-1):
        dd_tmp_list.append(compare.loc[compare.index[i+1]:,column].min()/compare.loc[compare.index[i],column]-1)
    return min(dd_tmp_list)
    


    
# 计算风险收益系列指标
def get_risk_indicators(compare,account_name,category,an_rate):
    '''
    
    风险收益指标计算

    INPUT：
    compare(df) 计算净值及沪深300全收益，
    account_name(str) 账户代码
    category(str) 账户类型，rela(相对收益),absol(绝对收益)
    an_rate(float) 绝对年化利率
    
    OUTPUT：
    rev_indicator(df) 收益指标,
    risk_indicator(df) 风险指标,
    ratio_indicator(df) 比率指标,
    year_summary(df) 年度展示,
    last_summary(df) 过去n月/年展示

    '''
    if category == 'rela':
        bench = 'hs300'
    elif category == 'absol':
        bench='absolute'
    else:
        print('Error...')
        
    account_chinese_name = gp_name.loc[gp_name['vc_gp3_account']==account_name,'vc_gp3_account_name'].values[0]

    ##年交易日数量
    days = 250

    ##基础风险指标
    #计算账户净值收益率
    compare['nv_return'] = compare['netvalue']/compare['netvalue'].shift(1)-1
    #计算账户净值波动率(年化),即日波动率乘以(250)^(1/2)
    std_an = np.std(compare['nv_return'])*(days)**(0.5)
    #计算超额收益率，进而计算年化跟踪误差
    compare['hs_return'] = compare[bench]/compare[bench].shift(1)-1
    compare['ex_return'] = compare['nv_return']-compare['hs_return']
    std_ex_an = np.std(compare['ex_return'])*(days)**(0.5)
    #计算最大回撤
    drawdown = get_drawdown(compare,'netvalue')
    #计算下行风险
    down_std_an = np.std(compare['nv_return'][compare['nv_return']<0])*(days)**(0.5)
    #计算下行风险
    down_std_an_b = np.std(compare['hs_return'][compare['hs_return']<0])*(days)**(0.5)
    #计算基准净值波动率(年化)
    std_b_an = np.std(compare['hs_return'])*(days)**(0.5)
    #计算基准最大回撤
    drawdown_b = get_drawdown(compare,bench)

    ##计算年化收益
    nv_return_an = (compare['netvalue'][compare.index[-1]]/compare['netvalue'][compare.index[0]])**(days/(len(compare)-1))-1
    if category == 'rela':
        hs_return_an = (compare[bench][compare.index[-1]]/compare[bench][compare.index[0]])**(days/(len(compare)-1))-1
    elif category == 'absol':
        hs_return_an = an_rate
    ex_return_an = nv_return_an-hs_return_an
    
    ##风险调整后收益
    #信息比率
    info_rate = (nv_return_an-hs_return_an)/std_ex_an
    #Sharpe/Sortino Ratio中的无风险收益为0
    sharpe = (nv_return_an-0)/std_an
    sortino = (nv_return_an-0)/down_std_an
    kumar = -(nv_return_an)/drawdown
    
    #基准指标
    sharpe_b = (hs_return_an-0)/std_b_an
    sortino_b = (hs_return_an-0)/down_std_an_b
    kumar_b = -(hs_return_an)/drawdown_b

    #计算累计收益
    nv_return_cum = compare['netvalue'][compare.index[-1]]/compare['netvalue'][compare.index[0]]-1
    hs_return_cum = compare[bench][compare.index[-1]]/compare[bench][compare.index[0]]-1
    ex_return_cum = nv_return_cum-hs_return_cum
    
    ##分年度展示
    period_year_data = compare.resample('Y','last')
    period_year_data = period_year_data.append(compare.loc[compare.index[0]])
    period_year_data = period_year_data.sort_index()
    period_year_data['nv_return_y'] = period_year_data['netvalue']/period_year_data['netvalue'].shift(1)-1
    period_year_data['hs_return_y'] = period_year_data[bench]/period_year_data[bench].shift(1)-1
    period_year_data['ex_return_y'] = period_year_data['nv_return_y']-period_year_data['hs_return_y']
    year_summary = period_year_data.loc[period_year_data.index[1]:,['nv_return_y','hs_return_y','ex_return_y']]
    year_summary.index = year_summary.index.year
    year_summary = year_summary.append(pd.DataFrame({'nv_return_y':[nv_return_an],'hs_return_y':[hs_return_an],'ex_return_y':[ex_return_an]},index=['年化']))
    year_summary = year_summary.append(pd.DataFrame({'nv_return_y':[nv_return_cum],'hs_return_y':[hs_return_cum],'ex_return_y':[ex_return_cum]},index=['累计']))    
    
    idx_order = list(year_summary.index[-2:])
    idx_order = idx_order+list(year_summary.index[:-2])
    year_summary = year_summary.reindex(idx_order)
    dict_tmp = {'ex_return_y':'超额收益','hs_return_y':'基准收益','nv_return_y':'账户收益'}
    col_list = list(year_summary.columns)
    col_list = [dict_tmp[i] for i in col_list]
    year_summary.columns = col_list
    year_summary = year_summary[['账户收益','基准收益','超额收益']]
    year_summary['最大回撤'] = [0 for i in range(len(year_summary))]
    year_summary['基准最大回撤'] = [0 for i in range(len(year_summary))]
    idx_order = list(year_summary.index[2:])
    for i in idx_order:
        year_summary.loc[i,'最大回撤'] = get_drawdown(compare[dt.datetime(i,1,1):dt.datetime(i,12,31)],'netvalue')
        year_summary.loc[i,'基准最大回撤'] = get_drawdown(compare[dt.datetime(i,1,1):dt.datetime(i,12,31)],bench)
        year_summary.loc[i,'年化波动率'] = np.std(compare.loc[dt.datetime(i,1,1):dt.datetime(i,12,31),'nv_return'])*(days)**(0.5)
        year_summary.loc[i,'基准年化波动率'] = np.std(compare.loc[dt.datetime(i,1,1):dt.datetime(i,12,31),'hs_return'])*(days)**(0.5)
        year_summary.loc[i,'年化跟踪误差'] = np.std(compare.loc[dt.datetime(i,1,1):dt.datetime(i,12,31),'ex_return'])*(days)**(0.5)
        year_summary.loc[i,'年化下行风险'] = np.std(compare.loc[dt.datetime(i,1,1):dt.datetime(i,12,31),'nv_return'][compare['nv_return']<0])*(days)**(0.5)
        year_summary.loc[i,'基准年化下行风险'] = np.std(compare.loc[dt.datetime(i,1,1):dt.datetime(i,12,31),'hs_return'][compare['hs_return']<0])*(days)**(0.5)
    
    year_summary = year_summary.fillna(0)

    
    year_summary.loc['累计','最大回撤'] = drawdown
    year_summary.loc['累计','基准最大回撤'] = drawdown_b
    year_summary.loc['累计','年化波动率'] = std_an
    year_summary.loc['累计','基准年化波动率'] = std_b_an
    year_summary.loc['累计','年化跟踪误差'] = std_ex_an
    year_summary.loc['累计','年化下行风险'] = down_std_an
    year_summary.loc['累计','基准年化下行风险'] = down_std_an_b
    year_summary['信息比率'] = year_summary['超额收益']/year_summary['年化跟踪误差']
    year_summary['夏普比率'] = year_summary['账户收益']/year_summary['年化波动率']
    year_summary['索提诺比率'] = year_summary['账户收益']/year_summary['年化下行风险']
    year_summary['卡玛比率'] = -year_summary['账户收益']/year_summary['最大回撤']
    year_summary['基准夏普比率'] = year_summary['基准收益']/year_summary['基准年化波动率']
    year_summary['基准索提诺比率'] = year_summary['基准收益']/year_summary['基准年化下行风险']
    year_summary['基准卡玛比率'] = -year_summary['基准收益']/year_summary['基准最大回撤']    
    year_summary.loc['年化','信息比率'] = info_rate#'-'
    year_summary.loc['年化','夏普比率'] = sharpe#'-'
    year_summary.loc['年化','索提诺比率'] = sortino#'-'
    year_summary.loc['年化','卡玛比率'] = kumar#'-'
    year_summary.loc['年化','基准夏普比率'] = sharpe_b#'-'
    year_summary.loc['年化','基准索提诺比率'] = sortino_b#'-'
    year_summary.loc['年化','基准卡玛比率'] = kumar_b#'-'
    year_summary[year_summary.columns[:10]] = year_summary[year_summary.columns[:10]].applymap(lambda x: str(format((x*100),'.2f'))+'%')
    year_summary[year_summary.columns[10:]] = year_summary[year_summary.columns[10:]].applymap(lambda x: format(x,'.2f'))

    year_summary.loc['年化','最大回撤'] = '-'
    year_summary.loc['年化','基准最大回撤'] = '-'
    year_summary.loc['年化','年化波动率'] = '-'
    year_summary.loc['年化','基准年化波动率'] = '-'
    year_summary.loc['年化','年化跟踪误差'] = '-'
    year_summary.loc['年化','年化下行风险'] = '-'
    year_summary.loc['年化','基准年化下行风险'] = '-'

    if category == 'absol':
        time_idx = pd.date_range(compare.index[0],compare.index[-1],freq='D')
        daily_rate = (1+an_rate)**(1/365)
        AbsoGrowth = pd.DataFrame({'AbsoGrowth':[1 for i in range(len(time_idx))]},index=time_idx)
        for i in range(len(AbsoGrowth)-1):
            AbsoGrowth.loc[AbsoGrowth.index[i+1],'AbsoGrowth'] = AbsoGrowth.loc[AbsoGrowth.index[i],'AbsoGrowth']*daily_rate    
        Abso_year_data = AbsoGrowth.resample('Y','last')
        Abso_year_data.loc[dt.datetime(2000,1,1)]=1
        Abso_year_data = Abso_year_data.sort_index()
        Abso_year_data = Abso_year_data/Abso_year_data.shift(1)
        Abso_year_data.index = Abso_year_data.index.year
        
        year_summary['基准最大回撤'] = '-' 
        year_summary['基准年化波动率'] = '-'
        year_summary['基准年化下行风险'] = '-'
        year_summary['年化跟踪误差'] = '-'
        year_summary['信息比率'] = '-'
        year_summary['基准夏普比率'] = '-'
        year_summary['基准索提诺比率'] = '-'
        year_summary['基准卡玛比率'] = '-'
        #year_summary.loc['年化','基准收益'] = str(format(100*an_rate,'.2f'))+'%'
        for k in Abso_year_data.index[1:]:
            year_summary.loc[k,'基准收益'] = str(format(100*(Abso_year_data.loc[k,'AbsoGrowth']-1),'.2f'))+'%'

    
    
    ##过去n月展示
    
    period_month_data = compare.resample('M','last')
    period_month_data = period_month_data.sort_index()
    if len(period_month_data)>13:
        nv_rev_past_3M = period_month_data.loc[period_month_data.index[-1],'netvalue']/period_month_data.loc[period_month_data.index[-4],'netvalue']-1
        nv_rev_past_6M = period_month_data.loc[period_month_data.index[-1],'netvalue']/period_month_data.loc[period_month_data.index[-7],'netvalue']-1
        nv_rev_past_1Y = period_month_data.loc[period_month_data.index[-1],'netvalue']/period_month_data.loc[period_month_data.index[-13],'netvalue']-1
        hs_rev_past_3M = period_month_data.loc[period_month_data.index[-1],bench]/period_month_data.loc[period_month_data.index[-4],bench]-1
        hs_rev_past_6M = period_month_data.loc[period_month_data.index[-1],bench]/period_month_data.loc[period_month_data.index[-7],bench]-1
        hs_rev_past_1Y = period_month_data.loc[period_month_data.index[-1],bench]/period_month_data.loc[period_month_data.index[-13],bench]-1
        ex_rev_past_3M = nv_rev_past_3M-hs_rev_past_3M 
        ex_rev_past_6M = nv_rev_past_6M-hs_rev_past_6M 
        ex_rev_past_1Y = nv_rev_past_1Y-hs_rev_past_1Y 

        append_month = pd.DataFrame({
            '超额收益':[ex_rev_past_3M,ex_rev_past_6M,ex_rev_past_1Y],
            '基准收益':[hs_rev_past_3M,hs_rev_past_6M,hs_rev_past_1Y],
            '账户收益':[nv_rev_past_3M,nv_rev_past_6M,nv_rev_past_1Y]
        },index=['-3M','-6M','-1Y']
        )

        append_month = append_month.applymap(lambda x: str(format((x*100),'.2f'))+'%')
        last_summary = year_summary.loc[['累计'],:]
        last_summary = last_summary.append(append_month)
    else:
        append_month = pd.DataFrame({
            '超额收益':['-','-'],
            '基准收益':['-','-'],
            '账户收益':['-','-']
        },index=['-6M','-1Y']
        )
        
        last_summary = year_summary.loc[['累计'],:]
        last_summary = last_summary.append(append_month)
    
    if len(period_month_data)>25:
        nv_rev_past_2Y = period_month_data.loc[period_month_data.index[-1],'netvalue']/period_month_data.loc[period_month_data.index[-25],'netvalue']-1
        hs_rev_past_2Y = period_month_data.loc[period_month_data.index[-1],bench]/period_month_data.loc[period_month_data.index[-25],bench]-1
        ex_rev_past_2Y = nv_rev_past_2Y-hs_rev_past_2Y
        append_month_2 = pd.DataFrame({
        '超额收益':[str(format((ex_rev_past_2Y*100),'.2f'))+'%'],
        '基准收益':[str(format((hs_rev_past_2Y*100),'.2f'))+'%'],
        '账户收益':[str(format((nv_rev_past_2Y*100),'.2f'))+'%']
        },index=['-2Y']
        )
        
    else:
        
        append_month_2 = pd.DataFrame({
            '超额收益':['-'],
            '基准收益':['-'],
            '账户收益':['-']
        },index=['-2Y']
        )
    
    last_summary = last_summary.append(append_month_2)
    last_summary.index.name = account_chinese_name 
    last_summary = last_summary[['账户收益','基准收益','超额收益']]
    
    rev_indicator = pd.DataFrame({
        '账户名称':[account_name],
        '账户名称(中文)':[account_chinese_name],
        '年化收益':[str(format((nv_return_an*100),'.2f'))+'%'],
        '年化基准收益':[str(format((hs_return_an*100),'.2f'))+'%'],
        '年化超额收益':[str(format(((ex_return_an)*100),'.2f'))+'%']  
    }
    )
    rev_indicator = rev_indicator[['账户名称','账户名称(中文)','年化收益','年化基准收益','年化超额收益']]

    risk_indicator = pd.DataFrame({
        '账户名称':[account_name],
        '账户名称(中文)':[account_chinese_name],
        '年化波动率':[str(format((std_an*100),'.2f'))+'%'],
        '年化跟踪误差':[str(format((std_ex_an*100),'.2f'))+'%'],
        '最大回撤':[str(format((drawdown*100),'.2f'))+'%'],
        '年化下行风险':[str(format((down_std_an*100),'.2f'))+'%'],
        '基准年化波动率':[str(format((std_b_an*100),'.2f'))+'%'],
        '基准最大回撤':[str(format((drawdown_b*100),'.2f'))+'%'],
    }
    )
    risk_indicator = risk_indicator[['账户名称','账户名称(中文)','年化波动率','基准年化波动率','年化跟踪误差','年化下行风险','最大回撤','基准最大回撤']]
    
    ratio_indicator = pd.DataFrame({
        '账户名称':[account_name],
        '账户名称(中文)':[account_chinese_name],
        '信息比率':[str(format(info_rate,'.2f'))],
        '夏普比率':[str(format(sharpe,'.2f'))],
        '索提诺比率':[str(format(sortino,'.2f'))],
        '卡玛比率':[str(format(kumar,'.2f'))],
        '基准夏普比率':[str(format(sharpe_b,'.2f'))],
        '基准索提诺比率':[str(format(sortino_b,'.2f'))],
        '基准卡玛比率':[str(format(kumar_b,'.2f'))]
    }
    )
    
    ratio_indicator = ratio_indicator[['账户名称','账户名称(中文)','信息比率','夏普比率','索提诺比率','卡玛比率','基准夏普比率','基准索提诺比率','基准卡玛比率']]
    
    if category == 'absol':
        risk_indicator['基准年化波动率']='-'
        risk_indicator['基准最大回撤']='-'
        ratio_indicator['基准夏普比率']='-'
        ratio_indicator['基准索提诺比率']='-'
        ratio_indicator['基准卡玛比率']='-'
        
    return rev_indicator,risk_indicator,ratio_indicator,year_summary,last_summary

def plot_year_summary(year_summary):
    '''
    
    画年度展示

    INPUT：
    year_summary(df) 年度展示
    
    OUTPUT：
    无返回值

    '''
    year_summary = year_summary[['账户收益','基准收益','超额收益']]
    year_summary = year_summary.applymap(lambda x: x[:-1])
    year_summary = year_summary.astype(float)
    year_summary.index = ['Annual','Cumulative']+list(year_summary.index[2:])
    dict_tmp = {'超额收益':'Excess','基准收益':'Benchmark','账户收益':'Portfolio'}
    col_list = list(year_summary.columns)
    col_list = [dict_tmp[i] for i in col_list]
    year_summary.columns = col_list
    year_summary.plot.bar()
    return None


def fancy_display(report_date,account_name,relaorabsol,target=30):
    '''
    
    净值系列综合展示

    INPUT：
    account_name(str) 账户中文名称，
    relaorabsol(dict) 账户类型，
    an_rate(float) 绝对年化利率
    
    OUTPUT：
    无返回值

    '''
    print(account_name)
    target=float(target)
    an_rate = (1+target/100)**(1/3)-1
    
    account_name = rela_gp[rela_gp['vc_gp3_account_name']==account_name]['vc_gp3_account'].values[0]
    #print(account_name)
    compare = get_netvalue(account_name,net_value,cash_rela,df_hs,an_rate)
    category = relaorabsol[account_name]
    if category == 'absol':
        print('\033[30;1m绝对收益账户-目标年化利率：'+str(format(100*an_rate,'.2f'))+'%\033[30;0m')
    elif category == 'rela':
        print('\033[30;1m相对收益账户\033[30;0m')
    print('--------------------')
    print('账户起始:',str(compare.index[0])[:10])
    print('报告期  :',str(report_date)[:10])
    print('--------------------')
    #20201029
    compare = compare[:report_date].copy()
    rev_indicator,risk_indicator,ratio_indicator,year_summary,last_summary = get_risk_indicators(compare,account_name,category,an_rate)
    print('收益指标')
    time.sleep(0.3)
    display(rev_indicator)
    print('风险指标')
    time.sleep(0.3)
    display(risk_indicator)
    print('风险调整后收益指标（年化）')
    time.sleep(0.3)
    display(ratio_indicator)
    print('分段综合展示')
    time.sleep(0.3)
    display(last_summary)
    print('年度综合展示')
    time.sleep(0.3)
    display(year_summary[year_summary.columns[:10]])
    display(year_summary[year_summary.columns[10:]])
    plot_year_summary(year_summary)
    
    
    if category == 'absol':
        plot_position(compare,an_rate,net_value,cash_rela,df_hs,df_stock,account_name)
    else:
        plot_net_value(compare,an_rate,category)
    
###计算区间收益_交互式
def get_period_return(compare,start_time,end_time):
    '''
    
    计算区间收益

    INPUT：
    compare(df) 计算净值及沪深300全收益，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    
    OUTPUT：
    nv_p_ret(float) 账户收益,
    hs_p_ret(float) 基准收益,
    ex_p_ret(float) 超额收益

    '''
    period_data = compare[start_time:end_time].copy()
    nv_p_ret = period_data.loc[period_data.index[-1],'netvalue']/period_data.loc[period_data.index[0],'netvalue']-1
    hs_p_ret = period_data.loc[period_data.index[-1],'hs300']/period_data.loc[period_data.index[0],'hs300']-1
    ex_p_ret = nv_p_ret-hs_p_ret
    return nv_p_ret,hs_p_ret,ex_p_ret

def display_period_return(account_name,start_time,end_time,net_value,cash_rela,df_hs,seperate,an_rate=0.0914):
    '''
    
    展示区间收益

    INPUT：
    account_name(str) 账户中文名称，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs(df) 沪深300数据
    
    OUTPUT：
    period_ret(df) 区间收益

    '''
    account_en_name = gp_name.loc[gp_name['vc_gp3_account_name']==account_name,'vc_gp3_account'].values[0]
    compare = get_netvalue(account_en_name,net_value,cash_rela,df_hs,an_rate)
    print('账户起始:',str(compare.index[0])[:10])
    time.sleep(0.3)
    try:
        if start_time < compare.index[0]:
            print('时间超出限制，起始至少应大于账户成立日',str(compare.index[0])[:10])
            period_ret = pd.DataFrame(columns=['#账户名称','#账户名称（中文）','#时间区间','区间收益','区间基准收益','区间超额收益'])
        else:
            nv_p_ret,hs_p_ret,ex_p_ret = get_period_return(compare,start_time,end_time)
            period_ret = pd.DataFrame({
                    '#账户名称':[account_en_name],
                    '#账户名称（中文）':[account_name],
                    '#时间区间':[str(start_time)[:11]+' - '+str(end_time)[:11]],
                    '区间收益':[str(format((nv_p_ret*100),'.2f'))+'%'],
                    '区间基准收益':[str(format((hs_p_ret*100),'.2f'))+'%'],
                    '区间超额收益':[str(format((ex_p_ret*100),'.2f'))+'%']
                }
                )

    except IndexError:
        print('时间超出限制，起始至少应小于结束日期')#,str(compare.index[0])[:10])
        period_ret = pd.DataFrame({
                '#账户名称':['-'],
                '#账户名称（中文）':['-'],
                '#时间区间':['-'],
                '区间收益':['-'],
                '区间基准收益':['-'],
                '区间超额收益':['-']
            }
            )
    seperate_copy = seperate.copy()
    seperate_copy['port_return'] = 0
    seperate_copy['excess_return'] = 0
    for i in range(len(seperate_copy)):
        if seperate_copy.loc[i,'end_time']<compare.index[0]:
            seperate_copy.loc[i,'port_return'] = '-'
            seperate_copy.loc[i,'bench_return'] = '-'
            seperate_copy.loc[i,'excess_return'] = '-'  
            seperate_copy.loc[i,'drawdown'] = '-'  
            seperate_copy.loc[i,'bench_drawdown'] = '-'  
            
        else:
            seperate_copy.loc[i,'port_return'] = round(100*get_period_return(compare,seperate_copy.loc[i,'start_time'],seperate_copy.loc[i,'end_time'])[0],2)
            seperate_copy.loc[i,'bench_return'] = round(100*get_period_return(compare,seperate_copy.loc[i,'start_time'],seperate_copy.loc[i,'end_time'])[1],2)
            seperate_copy.loc[i,'excess_return'] = (round(seperate_copy.loc[i,'port_return']-float(seperate_copy.loc[i,'bench_return']),2))
            seperate_copy.loc[i,'drawdown'] = (round(100*get_drawdown(compare[seperate_copy.loc[i,'start_time']:seperate_copy.loc[i,'end_time']],'netvalue'),2))
            seperate_copy.loc[i,'bench_drawdown'] = (round(100*get_drawdown(compare[seperate_copy.loc[i,'start_time']:seperate_copy.loc[i,'end_time']],'hs300'),2))
    
    
    

    #display(seperate)
    seperate_copy['port_return'] = seperate_copy['port_return'].map(lambda x:str('%.2f' %x)+'%' if x!='-' else '-')
    seperate_copy['excess_return'] = seperate_copy['excess_return'].map(lambda x:str('%.2f' %x)+'%' if x!='-' else '-')
    seperate_copy['bench_return'] = seperate_copy['bench_return'].map(lambda x:str('%.2f' %x)+'%' if x!='-' else '-')
    seperate_copy['drawdown'] = seperate_copy['drawdown'].map(lambda x:str('%.2f' %x)+'%' if x!='-' else '-')
    seperate_copy['bench_drawdown'] = seperate_copy['bench_drawdown'].map(lambda x:str('%.2f' %x)+'%' if x!='-' else '-')

    
    seperate_copy.columns = ['起始时间(固定)','结束时间(固定)','起始指数','结束指数','市场类型','持续期/月','基准收益','账户收益','超额收益','最大回撤','基准最大回撤']
    seperate_copy = seperate_copy[['起始时间(固定)','结束时间(固定)','市场类型','持续期/月','基准收益','账户收益','超额收益','最大回撤','基准最大回撤']]
    seperate_copy = seperate_copy.replace({'bull':'牛市','bear':'熊市','others':'震荡市'})
    display(seperate_copy)
    return period_ret

##交互式月度总结
def month_summary_template(period_month_data,name,cn_name):
    '''
    
    月度总结模板

    INPUT：
    period_month_data(df) 月度收益数据，
    name（str）指标名
    cn_name（str）指标中文名，
    
    OUTPUT：
    a(df) 月度总结模板

    '''
    if name == 'nv_month_ret':
        win_rate = str('%.2f' %round(sum(period_month_data[name]>=0)/(len(period_month_data[name])-1)*100,2))+'%'
    elif name == 'ex_month_ret_lag3':
        period_month_data_t = period_month_data[[name]].dropna(axis=0,how='any')
        try:
            win_rate = str('%.2f' %round(sum(period_month_data_t[name]>=0)/(len(period_month_data_t[name]))*100,2))+'%'
        except ZeroDivisionError:
            print('结束时间请至少比起始多3个月')
            time.sleep(0.2)
            win_rate = '-'
    elif name == 'ex_month_ret_lag6':
        period_month_data_t = period_month_data[[name]].dropna(axis=0,how='any')
        try:
            win_rate = str('%.2f' %round(sum(period_month_data_t[name]>=0)/(len(period_month_data_t[name]))*100,2))+'%'
        except ZeroDivisionError:
            print('结束时间请至少比起始多6个月')
            time.sleep(0.2)
            win_rate = '-'
    elif name == 'ex_month_ret_lag12':
        period_month_data_t = period_month_data[[name]].dropna(axis=0,how='any')
        try:
            win_rate = str('%.2f' %round(sum(period_month_data_t[name]>=0)/(len(period_month_data_t[name]))*100,2))+'%'
        except ZeroDivisionError:
            print('结束时间请至少比起始多12个月')
            time.sleep(0.2)
            win_rate = '-'
    else:
        win_rate = '-'
    a = pd.DataFrame({
        '##指标':[cn_name],
        '#平均值':[str('%.2f' %round(period_month_data[name].mean()*100,2))+'%'],
        '#最大值':[str('%.2f' %round(period_month_data[name].max()*100,2))+'%'],
        '#最小值':[str('%.2f' %round(period_month_data[name].min()*100,2))+'%'],
        '标准差':[str('%.2f' %round(period_month_data[name].std()*100,2))+'%'],
        '胜率':[win_rate]
    })
    return a
        
def get_month_summary(compare,start_time,end_time):
    '''
    
    计算月度总结

    INPUT：
    compare(df) 计算净值及沪深300全收益，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    
    OUTPUT：
    month_summary(df) 月度总结

    '''
    period_data = compare[start_time:end_time].copy()
    #display(period_data)
    period_month_data = period_data.resample('M','last')
    period_month_data = period_month_data.append(period_data.loc[period_data.index[0]])
    period_month_data = period_month_data.sort_index()
    period_month_data['hs_month_ret'] = period_month_data['hs300']/period_month_data['hs300'].shift(1)-1
    period_month_data['nv_month_ret'] = period_month_data['netvalue']/period_month_data['netvalue'].shift(1)-1
    period_month_data['nv_month_ret_lag3'] = period_month_data['netvalue']/period_month_data['netvalue'].shift(3)-1
    period_month_data['hs_month_ret_lag3'] = period_month_data['hs300']/period_month_data['hs300'].shift(3)-1
    period_month_data['nv_month_ret_lag6'] = period_month_data['netvalue']/period_month_data['netvalue'].shift(6)-1
    period_month_data['hs_month_ret_lag6'] = period_month_data['hs300']/period_month_data['hs300'].shift(6)-1
    period_month_data['nv_month_ret_lag12'] = period_month_data['netvalue']/period_month_data['netvalue'].shift(12)-1
    period_month_data['hs_month_ret_lag12'] = period_month_data['hs300']/period_month_data['hs300'].shift(12)-1

    period_month_data['ex_month_ret_lag3'] = period_month_data['nv_month_ret_lag3']-period_month_data['hs_month_ret_lag3']
    period_month_data['ex_month_ret_lag6'] = period_month_data['nv_month_ret_lag6']-period_month_data['hs_month_ret_lag6']
    period_month_data['ex_month_ret_lag12'] = period_month_data['nv_month_ret_lag12']-period_month_data['hs_month_ret_lag12']

    period_month_data['ex_month_ret'] = period_month_data['nv_month_ret']-period_month_data['hs_month_ret']
    
    from matplotlib.ticker import FuncFormatter
    
    def to_percent(temp,position):
        return '%1.0f'%(100*temp)+'%'
    
    hist = list(period_month_data['ex_month_ret'].dropna())
    plt.hist(hist,bins=int(len(hist)/3)+1)
    plt.title('Excess Return')
    plt.gca().xaxis.set_major_formatter(FuncFormatter(to_percent))
    month_summary = pd.DataFrame(columns=[ '##指标','#平均值','#最大值','#最小值','标准差','胜率'])
    month_summary = month_summary.append(month_summary_template(period_month_data,'nv_month_ret','月度收益'))
    month_summary = month_summary.append(month_summary_template(period_month_data,'hs_month_ret','基准收益'))
    month_summary = month_summary.append(month_summary_template(period_month_data,'ex_month_ret','超额收益'))
    month_summary = month_summary.append(month_summary_template(period_month_data,'ex_month_ret_lag3','滚动3月超额收益'))
    month_summary = month_summary.append(month_summary_template(period_month_data,'ex_month_ret_lag6','滚动6月超额收益'))
    month_summary = month_summary.append(month_summary_template(period_month_data,'ex_month_ret_lag12','滚动12月超额收益'))

    return month_summary 


def display_month_summary(account_name,start_time,end_time,net_value,cash_rela,df_hs,an_rate=0.0914):
    '''
    
    展示月度总结

    INPUT：
    account_name(str) 账户代码，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs(df) 沪深300数据
    
    OUTPUT：
    ms(df) 月度总结

    '''
    account_en_name = gp_name.loc[gp_name['vc_gp3_account_name']==account_name,'vc_gp3_account'].values[0]
    compare = get_netvalue(account_en_name,net_value,cash_rela,df_hs,an_rate)
    try:
        if start_time < compare.index[0]:
            print('时间超出限制，起始至少应大于账户成立日',str(compare.index[0])[:10])
            ms = pd.DataFrame(columns=[ '##指标','#平均值','#最大值','#最小值','标准差','胜率'])
        else:
            ms = get_month_summary(compare,start_time,end_time)
            ms.reset_index(drop=True,inplace=True)

    except IndexError:
        print('时间超出限制，起始至少应小于结束日期')
        time.sleep(0.2)
        ms = pd.DataFrame(columns=[ '##指标','#平均值','#最大值','#最小值','标准差','胜率'])
    return ms
```

```python

def get_netasset(account_name,net_value,cash_rela,df_hs):
    '''

    计算修复后净资产（部分账户现金流未及时计入净资产）

    INPUT：
    account_name(str) 账户代码，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    
    OUTPUT：
    tmp(df) 调整后净值

    '''
    tmp = net_value[net_value['vc_fund_code']==account_name]
    tmp.reset_index(drop=True,inplace=True)
    tmp.set_index('d_date',inplace=True)
    tmp = tmp.sort_index()
    tmp = tmp.reindex(df_hs.index)
    cash_tmp = cash_rela[cash_rela['vc_gp3_account']==account_name].copy()
    tmp = tmp.merge(cash_tmp,left_index=True,right_on='trade_dt',how='outer')
    tmp.drop('vc_gp3_account',axis=1,inplace=True)
    tmp['cashflow'] = tmp['cashflow'].fillna(0)
    tmp['raw_f_index_value'] = tmp['f_index_value']
    tmp.reset_index(drop=True,inplace=True)
    cash_index = list(tmp[tmp['cashflow']>0].index)
    for i in range(len(cash_index)):
        if np.isnan(tmp.loc[cash_index[i],'f_index_value']):
            tmp.loc[cash_index[i],'vc_fund_code'] = tmp.loc[cash_index[i]+1,'vc_fund_code']
            tmp.loc[cash_index[i],'vc_indicator_code'] = tmp.loc[cash_index[i]+1,'vc_indicator_code']
            tmp.loc[cash_index[i],'vc_indicator_name'] = tmp.loc[cash_index[i]+1,'vc_indicator_name']
            tmp.loc[cash_index[i],'f_index_value'] = 0
            tmp.loc[cash_index[i],'raw_f_index_value'] = 0

        #单独修复
        tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value'] = tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'f_index_value']+tmp.loc[(tmp['vc_fund_code']=='LH0025')&(tmp['trade_dt']==dt.datetime(2018,9,26)),'cashflow']
        if (tmp.loc[cash_index[i],'f_index_value'])/(tmp.loc[cash_index[i]+2,'f_index_value']+0.0001)<0.9:
            cash_missed = tmp.loc[cash_index[i],'cashflow']
            j =cash_index[i]
            #3-一般不会连续三天未增加
            while (tmp.loc[j,'f_index_value'])/(tmp.loc[j+2,'f_index_value']+0.0001)<0.9:
                #单独修复的账户
                if (tmp.loc[j,'vc_fund_code'] not in ['LH14G2','LH14H2','LH19I2'])|(tmp.loc[j,'trade_dt'] not in [dt.datetime(2020,9,23),dt.datetime(2020,9,24),dt.datetime(2018,10,16),dt.datetime(2019,4,15)]):
                    tmp.loc[j,'f_index_value']=tmp.loc[j,'f_index_value']+cash_missed
                j = j+1   
    tmp = tmp.dropna(axis=0,how='any')
    tmp.reset_index(drop=True,inplace=True)
    tmp = tmp[['trade_dt','vc_fund_code','f_index_value','vc_indicator_code','vc_indicator_name']].copy()
    tmp.columns = ['d_date','vc_fund_code','f_index_value','vc_indicator_code','vc_indicator_name']
    tmp = tmp.sort_values('d_date')
    
    if account_name in ['LH0026','LH19C2','LH19E2','LH19D2']:
        tmp  = tmp[tmp['d_date']>=dt.datetime(2019,5,16)]
    tmp.reset_index(drop=True,inplace=True)
    
    return tmp


def get_position(account_name,net_value,cash_rela,df_hs,df_stock):
    '''

    计算仓位

    INPUT：
    account_name(str) 账户代码，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs（df） 沪深300数据
    df_stock (df) 持仓数据
    
    OUTPUT：
    position[['d_date','position']](df) 逐日仓位

    '''
    
    net_asset_tmp = get_netasset(account_name,net_value,cash_rela,df_hs)
    holdings_tmp = df_stock[df_stock['vc_gp3_account']==account_name].groupby('d_date')[['f_marketvalue']].sum()
    holdings_tmp.reset_index(inplace=True)
    position = net_asset_tmp.merge(holdings_tmp,left_on='d_date',right_on='d_date',how='left')
    position = position.fillna(0)
    position['position'] = position['f_marketvalue']/position['f_index_value'] 
    return position[['d_date','position']]


###计算多账户区间多种指标
def get_period_index(compare,start_time,end_time):
    '''
    
    计算区间多种指标

    INPUT：
    compare(df) 计算净值及沪深300全收益，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    
    OUTPUT：
    nv_p_ret(float) 账户收益,
    hs_p_ret(float) 基准收益,
    ex_p_ret(float) 超额收益

    '''
    days=250
    period_data = compare[start_time:end_time].copy()
    #计算账户净值收益率
    period_data['nv_return'] = period_data['netvalue']/period_data['netvalue'].shift(1)-1
    #计算超额收益率，进而计算年化跟踪误差
    period_data['hs_return'] = period_data['hs300']/period_data['hs300'].shift(1)-1
    period_data['ex_return'] = period_data['nv_return']-period_data['hs_return']
    std_ex_an = np.std(period_data['ex_return'])*(days)**(0.5)
    std_an = np.std(period_data['nv_return'])*(days)**(0.5)
    nv_p_ret = period_data.loc[period_data.index[-1],'netvalue']/period_data.loc[period_data.index[0],'netvalue']-1
    hs_p_ret = period_data.loc[period_data.index[-1],'hs300']/period_data.loc[period_data.index[0],'hs300']-1
    ex_p_ret = nv_p_ret-hs_p_ret
    return std_ex_an,std_an,nv_p_ret,hs_p_ret,ex_p_ret

###展示多账户区间多种指标
def display_period_index(start_time,end_time,account_name_list,net_value,cash_rela,df_hs,an_rate=0.0914,submit=0):
    '''
    
    展示多账户区间指标

    INPUT：
    account_name_list(list) 账户中文名称list，
    start_time（datetime）起始时间，
    end_time（datetime）结束时间，
    net_value(df) 净值数据，
    cash_rela(df) 真实现金流数据，
    df_hs(df) 沪深300数据
    
    OUTPUT：
    period_ret(df) 区间收益

    '''
    if submit == '我已经选好了':
        period_ind = pd.DataFrame({
            '账户名称':[],
            '时间区间':[],
            '区间收益':[],
            '区间基准收益':[],
            '区间超额收益':[],
            '年化波动率':[],
            '年化跟踪误差':[]
        }
        )

        for account_name in account_name_list:
            account_en_name = gp_name.loc[gp_name['vc_gp3_account_name']==account_name,'vc_gp3_account'].values[0]
            compare = get_netvalue(account_en_name,net_value,cash_rela,df_hs,an_rate)
            try:
                if start_time < compare.index[0]:
                    period_ind_a = pd.DataFrame({
                        '账户名称':[account_name],
                        '时间区间':['起始时间超限，须大于账户成立日'],
                        '区间收益':['-'],
                        '区间基准收益':['-'],
                        '区间超额收益':['-'],
                        '年化波动率':['-'],
                        '年化跟踪误差':['-']
                        }
                        )
                else:
                    std_ex_an,std_an,nv_p_ret,hs_p_ret,ex_p_ret = get_period_index(compare,start_time,end_time)
                    period_ind_a = pd.DataFrame({
                            '账户名称':[account_name],
                            '时间区间':[str(start_time)[:11]+' - '+str(end_time)[:11]],
                            '区间收益':[str(format((nv_p_ret*100),'.2f'))+'%'],
                            '区间基准收益':[str(format((hs_p_ret*100),'.2f'))+'%'],
                            '区间超额收益':[str(format((ex_p_ret*100),'.2f'))+'%'],
                            '年化波动率':[str(format((std_an*100),'.2f'))+'%'],
                            '年化跟踪误差':[str(format((std_ex_an*100),'.2f'))+'%']
                        }
                        )

            except IndexError:
                period_ind_a = pd.DataFrame({
                    '账户名称':[account_name],
                    '时间区间':['结束时间超限，须小于账户成立日'],
                    '区间收益':['-'],
                    '区间基准收益':['-'],
                    '区间超额收益':['-'],
                    '年化波动率':['-'],
                    '年化跟踪误差':['-']
                    }
                    )
            period_ind = period_ind.append(period_ind_a)
            print('计算中，稍等...')
        period_ind = period_ind[['账户名称','时间区间','区间收益','区间基准收益','区间超额收益','年化波动率','年化跟踪误差']]
        period_ind.reset_index(drop=True,inplace=True)
        display(period_ind)
```

```python
def select_daily_stocktop10(date,account,df_stock):
    """
    输入日期及账户信息 返回该账户当天市值前十大个股
    
    date 日期
    account 账户
    df_stock 数据集
    
    """
    stock = df_stock[df_stock['d_date']==date].copy()
    stock1 = stock[stock['vc_gp3_account']==account]
    stock_top10 = stock1.sort_values(by = 'f_marketvalue',axis = 0,ascending = False)
    stock_top10 = stock_top10.head(10)
    return stock_top10

 
def select_daily_stockbottom10(date,account,df_stock):
    """
    输入日期及账户信息 返回该账户当天市值后十大个股
    
    date 日期
    account 账户
    df_stock 数据集
    
    """
    stock = df_stock[df_stock['d_date']==date].copy()
    stock1 = stock[stock['vc_gp3_account']==account]
    stock_bottom10 = stock1.sort_values(by = 'f_marketvalue',axis = 0,ascending = True)
    stock_bottom10 = stock_bottom10.head(10)
    return stock_bottom10


def daily_top10_holdrate(top10,date,account,df_stock):
    """
    输入期末top10个股 返回该账户某一天该十只个股的持仓比例
    
    top10 期末top10个股
    date 日期
    account 账户
    df_stock 数据集
    
    """
    stock = df_stock[df_stock['d_date']==date].copy()
    stock1 = stock[stock['vc_gp3_account']==account]    
    rate = []
    for symbol in top10['vc_symbol']:
        top10val = stock1[stock1['vc_symbol'] == symbol]['f_marketvalue'].values
        val = top10val/sum(stock1['f_marketvalue'])
        rate.append(val)
    for i in np.array(range(0,len(rate))):
        if len(rate[i]) == 0:
            rate[i] = np.array([0])
    rate = pd.Series(rate)
    rate = rate.astype(float)
    return rate

def select_daily_stock_effhd_num(date,account,df_stock):
    """
    输入日期及账户信息 返回该账户当天有效持股数量（市值大于100万）
    
    date 日期
    account 账户
    df_stock 数据集
    
    """
    stock = df_stock[df_stock['d_date']==date].copy()
    stock1 = stock[stock['vc_gp3_account']==account]
    stock_effhd = stock1[stock1['f_marketvalue'] > 1000000]
    return len(stock_effhd)

def stock_range_top10(date_start,date_end,account,df_stock):
    """
    输入一个时间段 返回该时期内该账户期间进入过前十大的所有股票 并记录其进入的日期
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_stock 数据集
    
    """
    time_range = pd.date_range(date_start,date_end)
    stocks = pd.DataFrame()
    for date in time_range:
        top10 = select_daily_stocktop10(date,account,df_stock)
        stocks = stocks.append(top10,ignore_index = True)
        stocks = stocks.sort_values(by = 'd_date',axis = 0,ascending = True)
        stocks.drop_duplicates(['vc_symbol'],keep = 'first',inplace = True)
    
    stocks.reset_index(drop=True,inplace=True)
    return stocks


def top10_day_into_alldays(date_top10,account,df_stock):
    """
    输入一个时间及账户信息 返回该期末该账户前十大股票第一次进入开始到输入日期为止的所有时间
    
    date_top10 期末时间
    account 账户
    df_stock 数据集
    
    """
    top10 = select_daily_stocktop10(date_top10,account,df_stock)
    time_range = pd.date_range(df_stock['d_date'].values[0],date_top10)
    stocks = pd.DataFrame()
    for date in time_range:
        top10_sgl = select_daily_stocktop10(date,account,df_stock)
        stocks = stocks.append(top10_sgl,ignore_index = True)
    stocks = stocks[stocks['vc_symbol'].isin(top10['vc_symbol'])]
    return stocks

#### 期末前十大个股持仓情况

def sheet_write_top10(date_top10,date1,date2,date3,account,df_stock,df_base):
    """
    作出期末前十大个股持仓汇总表 
    
    date_top10 期末日期
    date1 时期1
    date2 时期2
    date3 时期3
    account 账户
    df_stock 数据集
    df_base 基准数据集（一般选择沪深300）
    
    """
    ## 返回当期末与前3任意时期的持仓情况
    top10 = select_daily_stocktop10(date_top10,account,df_stock)
    val1 = daily_top10_holdrate(top10,date1,account,df_stock)
    val2 = daily_top10_holdrate(top10,date2,account,df_stock)
    val3 = daily_top10_holdrate(top10,date3,account,df_stock)
    val4 = daily_top10_holdrate(top10,date_top10,account,df_stock)
    ## 期间个股股价涨跌幅
    first_into = top10_day_into_alldays(date_top10,account,df_stock)
    first_into['d_date'] = first_into['d_date'].dt.date
    Time = []
    R = []
    for stock in top10['vc_full_name']:
        first = first_into[first_into['vc_full_name'] == stock]
        first = first.sort_values(by = 'd_date',axis = 0,ascending = True)
        first_start = first.head(1)
        first_end = first.tail(1)
        r =  float((first_end['f_close_marketvalue'].values-first_start['f_close_marketvalue'].values)/ first_start['f_close_marketvalue'].values)
        #display(first_start)
        #display(first_end)
        R.append(r)
        Time.append(first_start['d_date'].values)
    ## 期间沪深300全收益指数涨跌幅
    HS300 = []
    for stock in top10['vc_full_name']:
        first = first_into[first_into['vc_full_name'] == stock]
        hs300 = df_base[df_base['trade_dt'].isin(first['d_date'])]
        hs300 = hs300.sort_values(by = 'trade_dt',axis = 0,ascending = True)
        hs300_end = hs300.tail(1)
        hs300_start = hs300.head(1)
        hs_r = float((hs300_end['s_dq_close'].values-hs300_start['s_dq_close'].values)/ hs300_start['s_dq_close'].values)
        HS300.append(hs_r)
    ## 制表输出
    sheet = pd.DataFrame(top10['vc_full_name'].values,columns = ['个股名称/持仓比例'])
    sheet['进入前十大时间'] = Time
    sheet[date1] = val1
    sheet[date2] = val2
    sheet[date3] = val3
    sheet['当期末'] = val4
    sheet['期间个股涨跌幅'] = R
    sheet['期间沪深300全收益指数涨跌幅'] = HS300
    return sheet

#### 个股集中度走势
def sheet_write_top10_q(n,account,df_stock):
    """
    输入账户信息及数据集 返回前n个季度个股集中度走势 并绘图
    
    n 季度个数
    account 账户
    df_stock 数据集
    
    """
    df_stock = df_stock[df_stock['vc_gp3_account']==account]
    f3 = lambda x:x[-5:]
    df_stock['d_date'] = df_stock['d_date'].astype(str)
    df_stock["M_D"] = df_stock['d_date'].map(f3)
    df_stock_q = df_stock[df_stock["M_D"].isin(["03-31","06-30","09-30","12-31"])]
    ## 制表
    time = []
    val1 = []
    for date in df_stock_q['d_date']:
        top10 = select_daily_stocktop10(date,account,df_stock)
        val = sum(daily_top10_holdrate(top10,date,account,df_stock))
        val1.append(val)
        time.append(date)
    val2 = np.mean(val1)
    sheet = pd.DataFrame(time,columns = ['时间'])
    sheet['个股集中度（前十大）'] = val1
    sheet['平均'] = val2
    sheet.drop_duplicates(subset = None,keep = 'first',inplace = True)
    def MtoQ(x):
        MD = x[-5:]
        if MD =='03-31':
            val = str(x[0:4]+'Q1')
        if MD =='06-30':
            val = str(x[0:4]+'Q2')
        if MD =='09-30':
            val = str(x[0:4]+'Q3')
        if MD =='12-31':
            val = str(x[0:4]+'Q4')
        return val
    sheet['时间'] = sheet['时间'].map(MtoQ)
    sheet = sheet.tail(n)
    ## 绘图
    #plt.style.use('ggplot')
    plt.bar(sheet['时间'],sheet['个股集中度（前十大）'],color = 'royalblue',alpha = 0.8,label = 'Concentrations of individual stock')
    plt.plot(sheet['时间'],sheet['平均'],'r-',label = 'Average')
    plt.legend(bbox_to_anchor = (0.,1.,1.,.102),ncol = 2,mode = 'expend',borderaxespad = 0.)
    plt.show()  
    return sheet

#### 前N季度有效持股数量走势
def sheet_write_effhd_q(n,account,df_stock):
    """
    输入账户信息及数据集 返回前n个季度有效持股数量走势 并绘图
    
    n 季度个数
    account 账户
    df_stock 数据集
    
    """
    df_stock = df_stock[df_stock['vc_gp3_account']==account]
    f3 = lambda x:x[-5:]
    df_stock['d_date'] = df_stock['d_date'].astype(str)
    df_stock["M_D"] = df_stock['d_date'].map(f3)
    df_stock_q = df_stock[df_stock["M_D"].isin(["03-31","06-30","09-30","12-31"])]
    ## 制表
    time = []
    val1 = []
    for date in df_stock_q['d_date']:
        val = select_daily_stock_effhd_num(date,account,df_stock)
        val1.append(val)
        time.append(date)
    val2 = np.mean(val1)
    sheet = pd.DataFrame(time,columns = ['时间'])
    sheet['有效持股数量'] = val1
    sheet['平均'] = int(val2)
    sheet.drop_duplicates(subset = None,keep = 'first',inplace = True)
    def MtoQ(x):
        MD = x[-5:]
        if MD =='03-31':
            val = str(x[0:4]+'Q1')
        if MD =='06-30':
            val = str(x[0:4]+'Q2')
        if MD =='09-30':
            val = str(x[0:4]+'Q3')
        if MD =='12-31':
            val = str(x[0:4]+'Q4')
        return val
    sheet['时间'] = sheet['时间'].map(MtoQ)
    sheet = sheet.tail(n)
    ## 绘图
    #plt.style.use('ggplot')
    plt.bar(sheet['时间'],sheet['有效持股数量'],color = 'royalblue',alpha = 0.8,label = 'Effective number of shares')
    plt.plot(sheet['时间'],sheet['平均'],'r-',label = 'Average')
    plt.legend(bbox_to_anchor = (0.,1.,1.,.102),ncol = 2,mode = 'expend',borderaxespad = 0.)
    plt.show()  
    return sheet

#### 期间重仓前十个股走势和买卖操作情况
def range_top10_PB_ROE(date_start,date_end,account,df_stock,df_fin):
    """
    输入一个时间段 返回该时期内该账户所有进入过前十大股的PB与ROE
    
    date_start 起始时间
    date_end 结束时间
    account 账户代码
    df_stock 股票数据集
    df_fin 财务数据集
    
    """
    # 筛选年度报表
    f5 = lambda x:x[-5:]
    df_fin['report_period'] = df_fin['report_period'].astype(str)
    df_fin['M_D'] = df_fin['report_period'].map(f5)
    df_fin_year = df_fin[df_fin['M_D'] == '12-31']
    # 期间top10
    top10 = stock_range_top10(date_start,date_end,account,df_stock)
    df_fin_year_top10 = df_fin_year[df_fin_year['wind_code'].isin(top10['vc_symbol'])]
    # PB与ROE
    f6 = lambda x:x[0:4]
    df_fin_year_top10['Y'] = df_fin_year_top10['report_period'].map(f6)
    top10['d_date'] =  top10['d_date'].astype(str)
    top10['Y'] = top10['d_date'].map(f6)
    df_fin_year_top10['Y'] = df_fin_year_top10['Y'].astype(float)
    top10['Y'] = top10['Y'].astype(float)
    name = []
    PB = []
    ROE = []
    for stock in top10['vc_symbol']:
        top10_stock = top10[top10['vc_symbol'] == stock]
        fin = df_fin_year_top10[df_fin_year_top10['wind_code'] == stock]
        ### 判断该年年报是否已出 如未出则返回上一年年报数据
        if fin['Y'].values in (top10_stock['Y'].values):
            fin = fin[fin['Y'].isin(top10_stock['Y'])]
        else:
            fin = fin[fin['Y'].isin((top10_stock['Y']-1))]      
        pb = float(top10_stock['f_close_marketvalue'].values/fin['s_fa_bps'].values)
        roe = float(fin['s_fa_roe'].values)
        PB.append(pb)
        ROE.append(roe)
        name.append(top10_stock['vc_full_name'].values)
    # 制表输出
    sheet = pd.DataFrame(name,columns = ['重仓个股清单'])
    sheet['进入前十大时间'] = top10['d_date']
    sheet['PB'] = PB
    sheet['ROE'] = ROE
    # 绘制散点图
    #plt.style.use('ggplot')
    plt.scatter(sheet['PB'],sheet['ROE'],color = 'royalblue',alpha = 0.8)
    plt.xlabel('PB')
    plt.ylabel('ROE')
    plt.show()
    return sheet

#### 工具函数
def total_stock_profit(date_start,date_end,account,df_inv_stock):
    """
    输入一个时间段 计算该时期内该账户中股票各自的总收益
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_inv_stock 数据集
    
    """
    stock = df_inv_stock[df_inv_stock['vc_gp3_account']==account].copy()
    time_range = pd.date_range(date_start,date_end)
    stock1 = stock[stock['d_date'].isin(time_range)]
    stock_g = stock1.groupby('vc_full_name').sum()
    stock_g['total'] = stock_g['f_period_dividents'] + stock_g['f_period_afairchange'] + stock_g['f_period_tfairchange'] + stock_g['f_period_srdincome']
    return stock_g

def total_stock_daily_profit(date,account,df_inv_stock):
    """
    输入一个时期 计算该时期该账户中股票各自的总收益
    
    date 日期
    account 账户
    df_inv_stock 数据集
    
    """
    stock = df_inv_stock[df_inv_stock['vc_gp3_account']==account].copy()
    stock1 = stock[stock['d_date'] == date]
    stock_g = stock1.groupby('vc_full_name').sum()
    stock_g['total'] = stock_g['f_period_dividents'] + stock_g['f_period_afairchange'] + stock_g['f_period_tfairchange'] + stock_g['f_period_srdincome']
    return stock_g

def select_daily_inv_stocktop10(date_start,date_end,account,df_inv_stock):
    """
    输入一个时间段 返回该时期内该账户收益最高的前十大个股
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_inv_stock 数据集
    
    """
    stock = total_stock_profit(date_start,date_end,account,df_inv_stock)
    stock_top10 = stock.sort_values(by = 'total',axis = 0,ascending = False)
    stock_top10 = stock_top10.head(10)
    return stock_top10

def select_daily_inv_stockbottom10(date_start,date_end,account,df_inv_stock):
    """
    输入一个时间段 返回该时期内该账户收益最低（亏损最大）的前十大个股
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_inv_stock 数据集
    
    """
    stock = total_stock_profit(date_start,date_end,account,df_inv_stock)
    stock_bottom10 = stock.sort_values(by = 'total',axis = 0,ascending =True)
    stock_bottom10 = stock_bottom10.head(10)
    return stock_bottom10

def top10_inv_alldays(date_start,date_end,account,df_inv_stock):
    """
    输入一个时间段及账户信息 返回该时间段内该账户收益前十大股票第一次进入开始到该时间段末(或持有期末）的所有时间
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_inv_stock 数据集
    
    """
    top10 = select_daily_inv_stocktop10(date_start,date_end,account,df_inv_stock)
    top10['vc_full_name'] = top10.index
    time_range = pd.date_range(date_start,date_end)
    stocks = pd.DataFrame()
    for date in time_range:
        inv_stock = df_inv_stock[df_inv_stock['vc_full_name'].isin(top10['vc_full_name'])]
        """
        top10_sgl = inv_stock[inv_stock['d_date'] == date]
        top10_sgl['total'] =  top10_sgl['f_period_dividents'] + top10_sgl['f_period_afairchange'] +  top10_sgl['f_period_tfairchange'] +  top10_sgl['f_period_srdincome']
        if top10_sgl != []:"""
        stocks = stocks.append(inv_stock,ignore_index = True)  
    return stocks
#### 期间平均后十大股（暂缺收益贡献）

def sheet_write_bottom10_inv(date_start,date_end,account,df_stock,df_inv_stock):
    """
    输入一个时间段及账户信息 返回该时间段内收益后十大股票的持仓时间、价格涨跌及持仓占比情况
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_stock 持仓数据集
    df_inv_stock 收益数据集
       
    """
    # 十大个股持有的起始时间与结束时间
    bottom10 = select_daily_inv_stockbottom10(date_start,date_end,account,df_inv_stock)
    bottom10['vc_full_name'] = bottom10.index
    time_range = pd.date_range(date_start,date_end)
    bottom10_dt = df_inv_stock[df_inv_stock['vc_full_name'].isin(bottom10['vc_full_name'])]
    bottom10_dt_start = bottom10_dt[bottom10_dt['vc_gp3_account']==account]
    bottom10_dt_start = bottom10_dt_start[bottom10_dt_start['d_date'].isin(time_range)]
    bottom10_dt_start = bottom10_dt_start.sort_values(by = 'd_date',axis = 0,ascending = True)
    bottom10_dt_start.drop_duplicates(['vc_full_name'],keep = 'first',inplace = True)
    bottom10_dt_end = bottom10_dt[bottom10_dt['vc_gp3_account']==account]
    bottom10_dt_end = bottom10_dt_end[bottom10_dt_end['d_date'].isin(time_range)]
    bottom10_dt_end = bottom10_dt_end.sort_values(by = 'd_date',axis = 0,ascending = True)
    bottom10_dt_end.drop_duplicates(['vc_full_name'],keep = 'last',inplace = True)
    # 制表1
    bottom_summary = bottom10_dt_start.merge(bottom10_dt_end,left_on='vc_full_name',right_on='vc_full_name')
    sheet = pd.DataFrame(bottom_summary['vc_full_name'].values,columns = ['个股名称'])
    sheet['开始持有时间'] = bottom_summary['d_date_x'].values
    sheet['结束持有时间'] = bottom_summary['d_date_y'].values
    # 期间股价涨跌幅及占比
    bottom10_stocks = df_stock[df_stock['vc_gp3_account']==account]
    bottom10_stocks = bottom10_stocks[bottom10_stocks['vc_full_name'].isin(bottom10['vc_full_name'])]
    price = []
    rate = []
    averate = []
    for i in range(0,10):
        t_start = sheet['开始持有时间'][i]
        t_end = sheet['结束持有时间'][i]
        name = sheet['个股名称'][i]
        time_range = pd.date_range(t_start,t_end)
        stocks = bottom10_stocks[bottom10_stocks['d_date'].isin(time_range)]
        stocks = stocks[stocks['vc_full_name'] == name]
        stocks_start = stocks.head(1)
        stocks_end = stocks.tail(1)
        p = (stocks_end['f_close_marketvalue'].values-stocks_start['f_close_marketvalue'].values)/stocks_start['f_close_marketvalue'].values
        stocks_account1 = df_stock[df_stock['vc_gp3_account']==account]
        stocks_account = stocks_account1[stocks_account1['d_date'].isin(time_range)]
        ave = []
        tmp = list(set(stocks_account['d_date']))
        tmp.sort()
        #for date in stocks_account['d_date']:
        for date in tmp:
            stocks_per = stocks_account[stocks_account['d_date'] == date]
            ra = stocks_per[stocks_per['vc_full_name'] == name]['f_marketvalue'].values/sum(stocks_per['f_marketvalue'])
            if len(ra) == 0:
                ra = 0
            ave.append(ra)
        av = np.mean(ave) 
        stocks_account.drop_duplicates(['vc_full_name'],keep = 'last',inplace = True)
        r = stocks_end['f_marketvalue'].values/sum(stocks_account['f_marketvalue'])
        price.append(p)
        averate.append(av)
        rate.append(r)
    for i in np.array(range(0,len(price))):
        if len(price[i]) == 0:
            price[i] = np.array([0])
            print('%s 股票价格数据缺失'%sheet['个股名称'][i])
        if len(rate[i]) == 0:
            rate[i] =  np.array([0])
    price = pd.Series(price)
    price = price.astype(float)
    averate = pd.Series(averate)
    averate = averate.astype(float)
    rate = pd.Series(rate)
    rate = rate.astype(float)    
    # 制表2
    sheet['期间股价涨跌幅'] = price
    sheet['持有期间平均占比'] = averate
    sheet['持有期间期末占比'] = rate
    endrate = []
    sheet['结束持有时间'] = sheet['结束持有时间'].astype(str)
    for i in range(0,10):
        if sheet['结束持有时间'][i] == date_end:
            end = sheet['持有期间期末占比'][i]
        else:
            end = 0
        endrate.append(end)
    sheet['期末占比'] = endrate
    sheet['期间收益(亿元)'] = bottom10['total'].values/100000000
    sheet['期间收益(亿元)'] = sheet['期间收益(亿元)'].map(lambda x:format(x, '.2f'))
    return sheet

#### 期间平均前十大股（暂缺收益贡献）

def sheet_write_top10_inv(date_start,date_end,account,df_stock,df_inv_stock):
    """
    输入一个时间段及账户信息 返回该时间段内收益前十大股票的持仓时间、价格涨跌及持仓占比情况
    
    date_start 起始时间
    date_end 结束时间
    account 账户
    df_stock 持仓数据集
    df_inv_stock 收益数据集
       
    """
    # 十大个股持有的起始时间与结束时间
    top10 = select_daily_inv_stocktop10(date_start,date_end,account,df_inv_stock)
    top10['vc_full_name'] = top10.index
    time_range = pd.date_range(date_start,date_end)
    top10_dt = df_inv_stock[df_inv_stock['vc_full_name'].isin(top10['vc_full_name'])]
    top10_dt_start = top10_dt[top10_dt['vc_gp3_account']==account]
    top10_dt_start = top10_dt_start[top10_dt_start['d_date'].isin(time_range)]
    top10_dt_start = top10_dt_start.sort_values(by = 'd_date',axis = 0,ascending = True)
    top10_dt_start.drop_duplicates(['vc_full_name'],keep = 'first',inplace = True)
    top10_dt_end = top10_dt[top10_dt['vc_gp3_account']==account]
    top10_dt_end = top10_dt_end[top10_dt_end['d_date'].isin(time_range)]
    top10_dt_end = top10_dt_end.sort_values(by = 'd_date',axis = 0,ascending = True)
    top10_dt_end.drop_duplicates(['vc_full_name'],keep = 'last',inplace = True)
    # 制表1
    top_summary = top10_dt_start.merge(top10_dt_end,left_on='vc_full_name',right_on='vc_full_name')
    sheet = pd.DataFrame(top_summary['vc_full_name'].values,columns = ['个股名称'])
    sheet['开始持有时间'] = top_summary['d_date_x'].values
    sheet['结束持有时间'] = top_summary['d_date_y'].values
    # 期间股价涨跌幅及占比
    top10_stocks = df_stock[df_stock['vc_gp3_account']==account]
    top10_stocks = top10_stocks[top10_stocks['vc_full_name'].isin(top10['vc_full_name'])]
    price = []
    rate = []
    averate = []
    for i in range(0,10):
        t_start = sheet['开始持有时间'][i]
        t_end = sheet['结束持有时间'][i]
        name = sheet['个股名称'][i]
        time_range = pd.date_range(t_start,t_end)
        stocks = top10_stocks[top10_stocks['d_date'].isin(time_range)]
        stocks = stocks[stocks['vc_full_name'] == name]
        stocks_start = stocks.head(1)
        stocks_end = stocks.tail(1)
        p = (stocks_end['f_close_marketvalue'].values-stocks_start['f_close_marketvalue'].values)/stocks_start['f_close_marketvalue'].values
        stocks_account1 = df_stock[df_stock['vc_gp3_account']==account]
        stocks_account = stocks_account1[stocks_account1['d_date'].isin(time_range)]
        ave = []
        tmp = list(set(stocks_account['d_date']))
        tmp.sort()
        #for date in stocks_account['d_date']:
        for date in tmp:
            stocks_per = stocks_account[stocks_account['d_date'] == date]
            ra = stocks_per[stocks_per['vc_full_name'] == name]['f_marketvalue'].values/sum(stocks_per['f_marketvalue'])
            if len(ra) == 0:
                ra = 0
            ave.append(ra)
        av = np.mean(ave) 
        stocks_account.drop_duplicates(['vc_full_name'],keep = 'last',inplace = True)
        r = stocks_end['f_marketvalue'].values/sum(stocks_account['f_marketvalue'])
        price.append(p)
        averate.append(av)
        rate.append(r)
    for i in np.array(range(0,len(price))):
        if len(price[i]) == 0:
            price[i] = np.array([0])
            print('%s 股票价格数据缺失'%sheet['个股名称'][i])
        if len(rate[i]) == 0:
            rate[i] =  np.array([0])
    price = pd.Series(price)
    price = price.astype(float)
    averate = pd.Series(averate)
    averate = averate.astype(float)
    rate = pd.Series(rate)
    rate = rate.astype(float)    
    # 制表2
    sheet['期间股价涨跌幅'] = price
    sheet['持有期间平均占比'] = averate
    sheet['持有期间期末占比'] = rate
    endrate = []
    sheet['结束持有时间'] = sheet['结束持有时间'].astype(str)
    for i in range(0,10):
        if sheet['结束持有时间'][i] == date_end:
            end = sheet['持有期间期末占比'][i]
        else:
            end = 0
        endrate.append(end)
    sheet['期末占比'] = endrate
    sheet['期间收益(亿元)'] = top10['total'].values/100000000
    sheet['期间收益(亿元)'] = sheet['期间收益(亿元)'].map(lambda x:format(x, '.2f'))
    return sheet

#展示个股系列指标
def display_individual(end_time,time1,time2,time3,s_time_top,e_time_top,account_name,df_stock,df_hs300,df_fin,df_inv_stock):
    top10_stock = sheet_write_top10(end_time,time1,time2,time3,account_name,df_stock,df_hs300) 
    print('期末前十大个股')
    top10_stock[top10_stock.columns[2:]] = top10_stock[top10_stock.columns[2:]].astype(float)
    #display(top10_stock[top10_stock.columns[1:]].applymap(lambda x: format(x, '.2%')))
    top10_stock[top10_stock.columns[2:]] = top10_stock[top10_stock.columns[2:]].applymap(lambda x: format(x, '.2%')) 
    display(top10_stock)
    print('个股集中度走势')
    concentratrion_stock = sheet_write_top10_q(8,account_name,df_stock)
    concentratrion_stock[concentratrion_stock.columns[1:]] = concentratrion_stock[concentratrion_stock.columns[1:]].applymap(lambda x: format(x, '.2%')) 
    display(concentratrion_stock)
    print('有效持股数走势')
    eff_stock_num = sheet_write_effhd_q(8,account_name,df_stock)
    display(eff_stock_num)
    print('期间重仓（前十）个股估值情况')
    his_top10_pbroe = range_top10_PB_ROE(s_time_top,e_time_top,account_name,df_stock,df_fin)
    display(his_top10_pbroe)
    print('组合盈利前十个股')
    rev_top10 = sheet_write_top10_inv(s_time_top,e_time_top,account_name,df_stock,df_inv_stock)
    rev_top10[rev_top10.columns[3:7]] = rev_top10[rev_top10.columns[3:7]].applymap(lambda x: format(x, '.2%')) 
    display(rev_top10)
    print('组合亏损前十个股')
    rev_bottom10 = sheet_write_bottom10_inv(s_time_top,e_time_top,account_name,df_stock,df_inv_stock)
    rev_bottom10[rev_bottom10.columns[3:7]] = rev_bottom10[rev_bottom10.columns[3:7]].applymap(lambda x: format(x, '.2%')) 
    display(rev_bottom10)
    return 
```
